# The Logistic Map
> x = r.x.(1-x)

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

## Parameters
- Initial Population: x
- Growth Rate: r
- Number of recursive calls: i
- Population now: X
- Population later: Y

```python
x = 0.3
t = 0
X = []
Y = []
```

## Recursive Function Logistic Map

```python
def logisticMap(x,r,t): 
    if(t>1001):
        return -1
    X.append(x)
    x = r*x*(1-x)
    Y.append(x)
    t += 1
    return logisticMap(x,r,t)
```

```python
r = 1.0
logisticMap(x,r,t)
df1 = pd.DataFrame(list(zip(X, Y)), columns =['X(n)', 'X(n+1)'])
```

```python
r = 1.5
logisticMap(x,r,t)
df2 = pd.DataFrame(list(zip(X, Y)), columns =['X(n)', 'X(n+1)'])
```

```python
r = 2.0
logisticMap(x,r,t)
df3 = pd.DataFrame(list(zip(X, Y)), columns =['X(n)', 'X(n+1)'])
```

```python
r = 2.5
logisticMap(x,r,t)
df4 = pd.DataFrame(list(zip(X, Y)), columns =['X(n)', 'X(n+1)'])
```

```python
r = 3.0
logisticMap(x,r,t)
df5 = pd.DataFrame(list(zip(X, Y)), columns =['X(n)', 'X(n+1)'])
```

```python
r = 3.5
logisticMap(x,r,t)
df6 = pd.DataFrame(list(zip(X, Y)), columns =['X(n)', 'X(n+1)'])
```

```python
r = 4.0
logisticMap(x,r,t)
df7 = pd.DataFrame(list(zip(X, Y)), columns =['X(n)', 'X(n+1)'])
```

## Plotting X(n) vs. X(n+1) for different r values
- X(n) é uma taxa que varia no intervalo (0,1);
- A equação do mapa logístico representa uma parábola com a concavidade negativa, cujo vértice é representado cartesianamente pelo ponto (0.5, r/4), onde r é a taxa de crescimento;

```python
df1.plot(kind='scatter', x='X(n)', y='X(n+1)', color='red', label='r = 1.0')
df2.plot(kind='scatter', x='X(n)', y='X(n+1)', color='olive', label='r = 1.5')
df3.plot(kind='scatter', x='X(n)', y='X(n+1)', color='blue', label='r = 2.0')
df4.plot(kind='scatter', x='X(n)', y='X(n+1)', color='coral', label='r = 2.5')
df5.plot(kind='scatter', x='X(n)', y='X(n+1)', color='brown', label='r = 3.0')
df6.plot(kind='scatter', x='X(n)', y='X(n+1)', color='magenta', label='r = 3.5')
df7.plot(kind='scatter', x='X(n)', y='X(n+1)', color='green', label='r = 4.0')
legend = plt.legend(loc = 'upper left')
```
